### Feature Engineering

```python
tx_max_hire = data_before19.groupby('id_datorelavoro').rapporto_datainizio.max().reset_index()
tx_max_hire.columns = ['id_datorelavoro','latest_rapporto_datainizio']
tx_max_hire.head()
```

```python
#find the recency in days and add it to tx_user
tx_max_hire['difference_in_days'] = (tx_max_hire['latest_rapporto_datainizio'].max() - tx_max_hire['latest_rapporto_datainizio']).dt.days
data_user = pd.merge(data_user, tx_max_hire[['id_datorelavoro','difference_in_days']], on='id_datorelavoro')
data_user.head()
```

```python
#clustering for Recency
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(data_user[['difference_in_days']])
data_user['Cluster'] = kmeans.predict(data_user[['difference_in_days']])
```

```python
def order_cluster(cluster_field_name, target_field_name,df,ascending):
    new_cluster_field_name = 'new_' + cluster_field_name
    df_new = df.groupby(cluster_field_name)[target_field_name].mean().reset_index()
    df_new = df_new.sort_values(by=target_field_name,ascending=ascending).reset_index(drop=True)
    df_new['index'] = df_new.index
    df_final = pd.merge(df,df_new[[cluster_field_name,'index']], on=cluster_field_name)
    df_final = df_final.drop([cluster_field_name],axis=1)
    df_final = df_final.rename(columns={"index":cluster_field_name})
    return df_final
```

```python
data_user = order_cluster('Cluster', 'difference_in_days_y',data_user,False)
data_user.head()
```
